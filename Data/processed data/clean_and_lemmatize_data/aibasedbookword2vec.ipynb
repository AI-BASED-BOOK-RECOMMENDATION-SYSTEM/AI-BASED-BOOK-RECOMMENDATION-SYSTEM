```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# **Load the Data**

```python
# data = pd.read_csv("processed_data.csv",on_bad_lines='skip')
data = pd.read_csv("processed_data.csv",
                   on_bad_lines='skip',  # skip broken lines
                   quotechar='"',        # handle quoted text properly
                   encoding='utf-8')
```

```python
data.head()
```

```python
text=data['Title']+data['Author']+data['Genres']+data['Description']
```

```python
text
```

```python
data['text'] = data['Title'].fillna('') + ' ' + \
               data['Author'].fillna('') + ' ' + \
               data['Genres'].fillna('') + ' ' + \
               data['Description'].fillna('')
```

```python
data['text']
```

```python

import spacy
import pandas as pd


nlp = spacy.load("en_core_web_sm")



def preprocess_text(text):
    doc = nlp(text.lower())
    tokens = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
    return " ".join(tokens)


data['clean_text'] = data['text'].apply(preprocess_text)


print(data[['text','clean_text']].head())
```

```python
data.to_csv('clean_text.csv')
```
